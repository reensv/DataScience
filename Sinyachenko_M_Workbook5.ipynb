```python
#Task condition 1.1.4
#Создайте класс по работе с тригонометрическими функциями. В классе должны быть реализованы функции вычисления:
#− косинуса;
#− синуса;
#− тангенса;
#− арксинуса;
#− арккосинуса;
#− арктангенса;
#− перевода из градусов в радианы.

#solution
import math as m
import numpy as np
from termcolor import colored, cprint
print(colored('Введите тригонометрическую функцию', 'red', attrs=['underline']))
class trigonom:
    def sin(self, x):
        return np.sin(x)
    def cos(self, x):
        return np.cos(x)
    def tg(self, x):
        return np.tan(x)
    def arcsin(self, x):
        if -1 <= x <= 1:
            return np.arcsin(x)
        else:
            return "Ошибка"
    def arccos(self, x):
        if -1 <= x <= 1:
            return np.arccos(x)
        else:
            return "Ошибка"
    def arctg(self, x):
        return np.arctan(x)
    def deg_to_rad(self, x):
        return x * np.pi / 180

func = input()
print(colored('Введите значение аргумента функции', 'red', attrs=['underline']))
x = float(input())
if func == 'sin':
    print(trigonom().sin(x))
elif func == 'cos':
    print(trigonom().cos(x))
elif func == 'tan':
    print(trigonom().tg(x))
elif func == 'acos':
    print(trigonom().arccos(x))
elif func == 'atan':
    print(trigonom().arctg(x))
elif func == 'asin':
    print(trigonom().arcsin(x))
elif func == 'deg_to_rad':
    print(trigonom().geg_to_rad(x))
```

```python
#Task condition 1.2.3
#Представьте дерево показанное на рисунке с использованием списка из списков. Выведите на печать корень дерева, а также его левое и правое поддеревья.

#solution
class Tree:
    def __init__(self, kids, next=None):
        self.kids = self.val = kids
        self.next = next
        
t = Tree(Tree('a', Tree('b', Tree('d', 'e'))), Tree('c', Tree('f')))

print('root = ', t.kids.val)

print('left subtree = ', end='')
print(t.kids.next.val, end=', ')
print(t.kids.next.next.val, end=', ')
print(t.kids.next.next.next)

print('right subtree = ', end='')
print(t.next.kids, end=', ')
print(t.next.next.kids)
```

```python
#Task condition 1.2.4
#Дан класс, описывающий бинарное дерево.
#class Tree:
# def __init__(self, data):
# self.left = None
# self.right = None
# self.data = data
# def PrintTree(self):
# print(self.data)
#Реализуйте в классе функцию для вставки нового элемента в дерево по следующим правилам:
#• Левое поддерево узла содержит только узлы со значениями меньше, чем значение в узле.
#• Правое поддерево узла содержит только узлы со значениями меньше, чем значение в узле.
#• Каждое из левого и правого поддеревьев также должно быть бинарным деревом поиска.
#• Не должно быть повторяющихся узлов.
#Метод вставки сравнивает значение узла с родительским узлом и решает куда доваить элемент (в левое или правое поддерево). Перепишите, методPrintTree для печати полной версии дерева.

#solution
class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
    
    def PrintTree(self):
        if self.left:
            self.left.PrintTree()
        if self.right:
            self.right.PrintTree()
        print(self.data)
    
    def AddElement(self, Element):
        if self.data:
            if Element < self.data:
                if self.left is None:
                    self.left = Tree(Element)
                else:
                    self.left.AddElement(Element)
            elif Element > self.data:
                if self.right is None:
                    self.right = Tree(Element)
                else:
                    self.right.AddElement(Element)
        else:
            self.data = Element
```

```python
#Task condition 1.4.2
#Задание. Постройте модель регрессии для данных из предыдущей рабочей тетради.Для примера можно взять потребления газа (в миллионах галлонов) в 48 штатах США или набор данных о качестве красного вина:
#https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv
#https://raw.githubusercontent.com/aniruddhachoudhury/Red-WineQuality/master/winequality-red.csv
#Постройте прогноз. Оцените точность модели.

#solution
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import tree
from sklearn import metrics
url = 'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
dataframe = pd.read_csv(url)
dataframe.head()
dataframe.describe()

dataframe.plot(x="Petrol_tax",y=['Paved_Highways','Petrol_Consumption','Population_Driver_licence(%)'],kind="bar")
plt.xlabel("Штаты")
plt.ylabel("Потребление бензина")
plt.show()

X = dataframe[['Average_income','Population_Driver_licence(%)','Petrol_Consumption']]
y = dataframe['Petrol_tax']
print(X)
print(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)

from sklearn import tree
tree.plot_tree(regressor)

#прогноз:
y_pred = regressor.predict(X_test)
y_pred

df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df

print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))

metrics.mean_absolute_error(y_test, y_pred) / np.average(y) * 100
```
