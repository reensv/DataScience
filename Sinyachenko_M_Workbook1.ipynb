```python
#Task condition 4.3.2
#Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива – линейный и для среза – точечный.


#solution
import math as m
import numpy as np
import matplotlib.pyplot as plt
import random as r
import statistics as s
from scipy.integrate import simps
from numpy import trapz

a = []
f = []
for i in range(1, 11, 1):a.append(i)
    
for x in a:
    numerator = m.sqrt(1 + m.exp(m.sqrt(x)) + m.pow(m.cos(x), 2))
    denominator = abs(1 - m.pow(m.sin(x), 3))
    f.append(numerator/denominator)
plt.grid()
plt.plot(a, f, c = 'b')
arrX = [1, 2, 3, 4, 5]
plt.scatter(arrX, f[:5], c = 'r')
```

```python
#Task condition 4.3.3
#Дана функция. Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним. 

#solution
import math as m
import numpy as np
import matplotlib.pyplot as plt
import random as r
import statistics as s
from scipy.integrate import simps
from numpy import trapz

a = []
f = []
for i in range(10): a.append(i)
for x in a:
    f.append(abs(m.cos(x*m.exp(m.cos(x) + np.log(x+1)))))
plt.grid()
plt.plot(a, f, c = 'r')
plt.fill_between(a, f)
area = trapz(f)
print(area)
```
