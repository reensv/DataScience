```python
#Task condition 4.3.2
#–°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì –ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π.


#solution
import math as m
import numpy as np
import matplotlib.pyplot as plt
import random as r
import statistics as s
from scipy.integrate import simps
from numpy import trapz

a = []
f = []
for i in range(1, 11, 1):a.append(i)
    
for x in a:
    numerator = m.sqrt(1 + m.exp(m.sqrt(x)) + m.pow(m.cos(x), 2))
    denominator = abs(1 - m.pow(m.sin(x), 3))
    f.append(numerator/denominator)
plt.grid()
plt.plot(a, f, c = 'b')
arrX = [1, 2, 3, 4, 5]
plt.scatter(arrX, f[:5], c = 'r')
```

```python
#Task condition 4.3.5
#–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä.

#solution
import math as m
import numpy as np
import matplotlib.pyplot as plt
import random as r
import statistics as s
from scipy.integrate import simps
from numpy import trapz
from termcolor import colored, cprint

def addition(x, y):
    return x + y

def subtraction(x, y):
    return x - y

def multiplication(x, y):
    return x*y
    
def division(x, y):
    return x/y 

def exponentiation_e(x, e):
    return m.exp(x+y)

def sine(x, y):
    return m.sin(x + y)
    
def cosine(x, y):
    return m.cos(x + y)
    
def exponentiation(x, y):
    return m.pow(x, y)

print(colored('–í—ã–±–µ—Ä–∏—Ç–µ —Ç–∏–ø –æ–ø–µ—Ä–∞—Ü–∏–∏', 'red', attrs=['underline']))
print(colored('+', 'red'))
print(colored('-', 'red'))
print(colored('*', 'red'))
print(colored('/', 'red'))
print(colored('–¥—Ä—É–≥–æ–µ', 'blue'))
a = input()
if a == '–¥—Ä—É–≥–æ–µ':
    print(colored('–≤–æ–∑–≤–µ—Å—Ç–∏ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É –≤ —Å—Ç–µ–ø–µ–Ω—å —Å—É–º–º—ã —á–∏—Å–µ–ª', 'blue'))
    print(colored('—Å–∏–Ω—É—Å —Å—É–º–º—ã', 'blue'))
    print(colored('–∫–æ—Å–∏–Ω—É—Å —Å—É–º–º—ã', 'blue'))
    print(colored('–≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å', 'blue'))
    a = input()
x = float(input())   
y = float(input()) 

if a == '+':
    print(addition(x, y))
elif a == '-':
    print(subtraction(x, y))
elif a == '*':
    print(multiplication(x, y))
elif a == '/':
    print(division(x, y))
elif a == '–≤–æ–∑–≤–µ—Å—Ç–∏ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ç—É –≤ —Å—Ç–µ–ø–µ–Ω—å —Å—É–º–º—ã —á–∏—Å–µ–ª':
    print(exponentiation_e(x, e))
elif a == '—Å–∏–Ω—É—Å —Å—É–º–º—ã':
    print(sine(x, y))
elif a == '–∫–æ—Å–∏–Ω—É—Å —Å—É–º–º—ã':
    print(cosine(x, y))
elif a == '–≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å':
    print(exponentiation(x, y))
```
